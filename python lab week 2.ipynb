```python
# Program: Arithmetic operations

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print ('Sum: ', a + b)
print ('Subtraction: ', a - b)
print ('Multiplication: ', a * b)
print ('Division: ', a / b)
print ('Floor Division: ', a // b)
print ('Modulo: ', a % b)
print ('Power: ', a ** b)
```

```python
x = int(input("Enter first number: "))
y = int(input("Enter second number: "))
a1= x & y #Bitwise AND
a2= x | y #Bitwise OR
a3= ~x #Bitwise NOT
a4=x ^ y # Bitwise XOR
a5=x >> 2 # >> Bitwise right shift
a6=x<<2 #<<Bitwise left shift
print( "bitwise AND=",a1)
print( "bitwise OR=",a2)
print( "bitwise NOT=",a3)
print( "bitwise XOR=",a4)
print( "bitwise Right shift=",a5)
print( "bitwise Left shift=",a6)
```

```python
# Program: Assignment operators
x = 10
print("Initial value:", x)

x += 5   # x = x + 5
print("After += :", x)

x -= 3   # x = x - 3
print("After -= :", x)

x *= 2   # x = x * 2
print("After *= :", x)

x /= 4   # x = x / 4
print("After /= :", x)

x //= 2  # floor division
print("After //= :", x)

x %= 3   # modulus
print("After %= :", x)
```

```python
# Program to check membership in a list

numbers = [10, 20, 30, 40, 50]

# Check if 20 is in the list
print("Is 20 in list?", 20 in numbers)

# Check if 60 is not in the list
print("Is 60 not in list?", 60 not in numbers)
```

```python
# Program to check membership in a string

text = "Python Programming"

print("Is 'P' in text?", 'P' in text)
print("Is 'Java' in text?", 'Java' not in text)
```
